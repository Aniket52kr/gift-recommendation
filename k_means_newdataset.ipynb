```python
# function to create a dictionary of unique values of column mapped to numerical values
def getMappingDictionary(unique_data):
    mappingDictionary = {}
    for i in range(len(unique_data)):
        mappingDictionary[unique_data[i]] = i
    return mappingDictionary
```

```python
def getUniqueDataList(columnName):
    return list(set(dataset[columnName]))
```

```python
def modifyDatasetColumn(columnName, mapping_dict):
    dataset[columnName] = dataset[columnName].map(mapping_dict)
```

```python
def modifyDataset(columnName):
    unique_columnValues = getUniqueDataList(columnName)
    print(unique_columnValues)
    mapping_dict = getMappingDictionary(unique_columnValues)
    print(mapping_dict)
    modifyDatasetColumn(columnName,mapping_dict)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import pandas as pd

# Assume df is your dataframe
# Apply one-hot encoding to all categorical columns
df_encoded = pd.get_dummies(df, drop_first=True)

# Create a KMeans model and use the KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1, 10))

# Fit the data to the visualizer
visualizer.fit(df_encoded)

# Display the elbow plot
visualizer.show()

# Get the optimal number of clusters (elbow value)
k = visualizer.elbow_value_
print("Optimal number of clusters:", k)
```

```python
# Initialize KMeans model
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)

# Fit the KMeans model and predict cluster labels
customer_segments = kmeans.fit_predict(df_encoded)

# Add cluster labels to the dataframe
df_encoded['cluster'] = customer_segments
```

```python
# Sample input data for the new customer
input_customer_data = {
    'age': [40],
    'gender': ['Female'],  # Raw value, needs encoding
    'relationship': ['Colleague'],  # Raw value, needs encoding
    'occasion': ['Karva Chauth'],  # Raw value, needs encoding
    'budget': [10000],  # Numerical value
    'interests': ['Photography,Astronomy']  # String of interests, needs splitting or encoding
}

# Convert new customer data to DataFrame
new_customer = pd.DataFrame(input_customer_data)

# Debug: Print the column names of the new customer DataFrame to check for 'interests'
print("Columns in new_customer DataFrame:", new_customer.columns)

# Check if 'interests' column exists
if 'interests' in new_customer.columns:
    # Example: Split the 'interests' column if needed (splitting interests into separate columns or one-hot encoding)
    interests_list = new_customer['interests'].str.split(',', expand=True)  # Split the interests by comma
    interests_list.columns = ['Photography', 'Astronomy']  # Rename the columns
    new_customer = pd.concat([new_customer, interests_list], axis=1)  # Concatenate to the original DataFrame
    new_customer = new_customer.drop('interests', axis=1)  # Drop the original 'interests' column

    # Debug: Print the DataFrame after handling the 'interests' column
    print("DataFrame after splitting 'interests' column:\n", new_customer)
else:
    print("No 'interests' column found in the new customer data.")

# Encode the categorical columns for new customer data
new_customer_encoded = pd.get_dummies(new_customer, drop_first=True)

# Debug: Print the first few rows of the encoded DataFrame
print("Encoded new_customer DataFrame:\n", new_customer_encoded.head())

# Make sure the new customer data has the same columns as the training data (after encoding)
# You need to ensure the columns align with the ones used in your training dataset.
new_customer_encoded = new_customer_encoded[['age', 'budget', 'interests', 'gender',  'occasion']]  # Example columns

# Now you can predict the cluster for the new customer
new_customer_segment = kmeans.predict(new_customer_encoded)

# Print the predicted segment
print('New customer belongs to segment:', new_customer_segment[0])
```
